```python
import pandas as pd

df = pd.read_csv("student_dropout_dataset_v3 (1).csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Dropout'].value_counts()
```

```python
import matplotlib.pyplot as plt

df['Dropout'].value_counts().plot(kind='bar')
plt.title("Dropout Distribution")
plt.xlabel("Dropout")
plt.ylabel("Count")
plt.show()
```

```python
import seaborn as sns

sns.histplot(df['Age'], bins=20)
plt.title("Age Distribution")
plt.show()
```

```python
sns.histplot(df['GPA'], bins=20)
plt.title("GPA Distribution")
plt.show()
```

```python
sns.histplot(df['Attendance_Rate'], bins=20)
plt.title("Attendance Rate Distribution")
plt.show()
```

```python
sns.boxplot(x='Dropout', y='GPA', data=df)
plt.title("GPA vs Dropout")
plt.show()
```

```python
sns.boxplot(x='Dropout', y='Attendance_Rate', data=df)
plt.title("Attendance vs Dropout")
plt.show()
```

```python
sns.boxplot(x='Dropout', y='Study_Hours_per_Day', data=df)
plt.title("Study Hours vs Dropout")
plt.show()
```

```python
sns.boxplot(x='Dropout', y='Stress_Index', data=df)
plt.title("Stress Index vs Dropout")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
pd.crosstab(df['Gender'], df['Dropout'])
```

```python
sns.countplot(x='Gender', hue='Dropout', data=df)
plt.title("Gender vs Dropout")
plt.show()
```
